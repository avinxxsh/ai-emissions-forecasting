```python
# imputing null values of remaining columns with manageable number of null values using median of category and subcategory
df['yearly_tec'] = df.groupby(['category', 'subcategory'])['yearly_tec'] \
                     .transform(lambda x: x.fillna(x.median()))

df['gwp_manufacturing_ratio'] = df.groupby(['category', 'subcategory'])['gwp_manufacturing_ratio'] \
                     .transform(lambda x: x.fillna(x.median()))

df['gwp_error_ratio'] = df.groupby(['category', 'subcategory'])['gwp_error_ratio'] \
                     .transform(lambda x: x.fillna(x.median()))

df['weight'] = df.groupby(['category', 'subcategory'])['weight'] \
                     .transform(lambda x: x.fillna(x.median()))

df['screen_size'] = df.groupby(['category', 'subcategory'])['screen_size'] \
                     .transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df['added_date'] = pd.to_datetime(df['added_date'], format='mixed', errors='coerce')
df['report_date'] = pd.to_datetime(df['report_date'], format='mixed', errors='coerce')
```

```python
df['report_year'] = df['report_date'].dt.year
```

```python
# derived columns
df['annual_lifecycle'] = df['gwp_total'] / df['lifetime'] # ghg emissions per year per product 
df['gwp_use'] = df['gwp_total'] * df['gwp_use_ratio'] # use phase emissions
df['gwp_manufacturing'] = df['gwp_total'] * df['gwp_manufacturing_ratio']
df["use_share"] = df["gwp_use"] / df["gwp_total"]
df["manuf_share"] = df["gwp_manufacturing"] / df["gwp_total"]
```

```python
server_df = df_dc[
    (df_dc["gwp_total"] < 30000) &
    (df_dc["yearly_tec"] < 5000)
].copy()

server_df.describe()
```

```python
server_df.shape
```

```python
server_df["report_year"].value_counts().sort_index()
```

```python
# building historical trend dataset
trend = (
    server_df.groupby('report_year')
    .agg({
        'gwp_total': 'median',
        'annual_lifecycle': 'median',
        'yearly_tec': 'median',
        'gwp_use_ratio': 'median',
        'gwp_manufacturing_ratio': 'median',
        'gwp_use': 'median',
        'gwp_manufacturing': 'median'
    })
    .reset_index()
)

trend
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# because only 1 data point per year from 2016 - 2018
trend_clean = trend[trend['report_year'].isin([2019, 2020, 2021])]

X = trend_clean['report_year'].values.reshape(-1, 1)
y = trend_clean['annual_lifecycle'].values

lin_model = LinearRegression()
lin_model.fit(X, y)

exp_model = LinearRegression()
exp_model.fit(X, np.log(y))
```

```python
# projections
future_years = np.arange(2022, 2027).reshape(-1, 1)

linear_pred = lin_model.predict(future_years)
exp_pred = np.exp(exp_model.predict(future_years))
```

```python
projection = pd.DataFrame({
    "year": future_years.flatten(),
    "linear_proj": linear_pred,
    "exp_proj": exp_pred
})
```

```python
historical = trend_clean[["report_year", "annual_lifecycle"]]
historical.columns = ["year", "annual_lifecycle"]

final_proj = pd.concat([historical, projection], ignore_index=True)
```

```python
plt.figure(figsize=(10,6))

plt.plot(historical['year'], historical['annual_lifecycle'], 'o-', label='Historical')
plt.plot(projection['year'], projection['linear_proj'], 's-', label='Linear Projection')
plt.plot(projection['year'], projection['exp_proj'], '^-', label='Exponential Projection')

plt.xlabel("Year")
plt.ylabel("Annual Lifecycle Emissions (kgCO2e per server)")
plt.title("AI Server Emissions: Historical Trend and 5-Year Projection")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# cloud server electricity estimate for comparision
typ_server = {
    "yearly_tec_kwh": df_dc['yearly_tec'].median() 
}

typ_server
```

```python
n_servers = 10000000 # estimate
cloud_electricity_kwh = typ_server['yearly_tec_kwh'] * n_servers
cloud_electricity_twh = cloud_electricity_kwh / 1e9 # to match owid dataset units

cloud_electricity_twh
```

```python
owid = pd.read_csv('owid-energy-data.csv')
owid.head()
```

```python
year = 2021
country = "Europe"

mask = (owid['year'] == year) & (owid['country'] == country)
elec = owid.loc[mask, ['country', 'electricity_demand']].dropna()

elec
```

```python
comparision = elec.copy()
comparision = comparision.rename(columns={'electricity_demand': 'electricity_twh'})

cloud_row = pd.DataFrame({
    'country': ['Hypothetical Sample (10M Servers)'],
    'electricity_twh': [cloud_electricity_twh]
})

comparision = pd.concat([comparision, cloud_row], ignore_index = True)
comparision
```

```python
plt.figure(figsize=(8, 5))
plt.bar(comparision['country'], comparision['electricity_twh'])
#plt.xticks(ha='right')
plt.ylabel('Electricity demand (TWh per year)')
plt.title(f'Electricity use: Cloud scenario vs Europe, {year}')
plt.tight_layout()
plt.show()
```
